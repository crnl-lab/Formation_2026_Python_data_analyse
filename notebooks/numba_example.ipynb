```python
import numba
import numpy as np
import matplotlib.pyplot as plt
```

```python


rng = np.random.default_rng(seed=2205)
data = rng.normal(size=(1_000_000))
data.shape
```

```python

def my_function(data):
    data2 = data.copy()
    for i in range(data.shape[0]):
        v = data[i]
        if i % 5 == 0 and v > 0.2:
            data2[i] = v**2
        else:
            data2[i] = -v**2 + 5.
    return data2

```

```python
%timeit my_function(data)
```

```python

@numba.jit(nopython=True, parallel=False)
def my_function_fast(data):
    data2 = data.copy()
    for i in range(data.shape[0]):
        v = data[i]
        if i % 5 == 0 and v > 0.2:
            data2[i] = v**2
        else:
            data2[i] = -v**2 + 5.
    return data2
```

```python
%timeit my_function_fast(data)
```

```python
@numba.jit(nopython=True, parallel=True)
def my_function_super_fast(data):
    data2 = data.copy()
    for i in numba.prange(data.shape[0]):
        v = data[i]
        if i % 5 == 0 and v > 0.2:
            data2[i] = v**2
        else:
            data2[i] = -v**2 + 5.
    return data2
```

```python
%timeit my_function_super_fast(data)
```
