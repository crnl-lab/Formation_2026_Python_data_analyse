## Création d'une fonction

On peut créer des fonctions avec des variables en entrée et une variable de sortie.

Attention à l'indentation, cela definit le **block**.

En python c'est 4 espaces.



```python
def mon_add(a, b):
    j = a+b
    s = f'Résultat: {a} + {b} = {j}'
    return s    

res = mon_add(5, 4)
print(res)
#print(j)
```

```python
aa = 3
bb = 77
res = mon_add(aa, bb)
print(res)
```

Il peut y avoir plusieurs variables de sortie.

```python
def mon_add_bis(a, b):
    c = a+b
    s = f'Résultat: {a} + {b} = {c}'
    return s, c

res, res2 = mon_add_bis(5, 4)
print(res)
print(res2)
```

Python permet d'avoir des variables facultatives.

```python
def mon_add_ter(a, b, c=1, d=3):
    t = a+b+c+d
    s = f'Résultat: {a} + {b} + {c} + {d} = {t}'
    return s

res = mon_add_ter(aa, bb)
print(res)
res = mon_add_ter(aa, bb, c = 6, d = 9)
print(res)
res = mon_add_ter(aa, bb, d = 9)
print(res)
```

Voir uniquement des varibales facultatives.

```python
def mon_add_4(c = 1, d = 3):
    t = c+d
    s = f'Résultat: {c} + {d} = {t}'
    return s

res = mon_add_4() # ici y'a rien
print(res)
```

```python
liste_de_lettres = ['a','b','c','d','e','f','g','h']

for lettre in liste_de_lettres:
    print(lettre)
```

### Range avec plusieurs arguments

L'aide de range propose plusieurs signatures :

- La signature avec un seul argument que nous avons utilisée jusqu'ici : `range(stop)`.
- Une signature avec 2 arguments : `range(start, stop)`
- Une signature avec 3 arguments : `range(start, stop, step)`

```python
range?
```

```python
for i in range(5,10):
    print(i)
```

```python
for i in range(5,20,2):
    print(i)
```

```python
for i in range(20,2,-1):
    print(i)
```
