### View and copy

Un tableau issu d'un même tableau est une *'view'*.


```python
a = np.arange(10)
a
```

```python
a2 = a[2:-2]
print(a2)
a2[0:6] = np.arange(6)*-10
print(a2)
print(a)
```

Sinon il faut faire une copie explicite.

```python
a = np.arange(10)
a2 = a[2:-2].copy()
a2[0:6] = np.arange(6)*-10
print(a2)
print(a)
```

### Concatenation de tableau

```python
a = np.arange(3)
b = np.arange(5)

c = np.concatenate((a,b)) ### concatene par défault selon la premiere dimension

print(c)
```

```python
### en multi dimension, on precise axis

a = np.ones((3,4,5))
b = np.ones((3,4,5))

c = np.concatenate((a,b),axis = 0) ### concatene selon la premiere dimension

print(c.shape)
```

![Image](img/np_concat_0.png)

```python
### Si les dimensions (autre que celle ou l'on concatène) ne correspondent pas : ça ne marche pas !

a = np.ones((3,4,5))
b = np.ones((3,5,3))

c = np.concatenate((a,b),axis = 1) 

print(c.shape)
```

# Operations de réduction: mean/std, max,  sum etc.

Avec le meme principe de 'axis', il est tres facile de reduire un tableau:

![Image](img/axis.png)

```python
rnd = np.random.default_rng(seed=2406)
m = rnd.integers(size=(3, 5), low=0, high=10)


print(m)

##Petit truc pour connaitre l'axis: c'est la dimension qui disparait

print()
mean_m = np.mean(m,axis = 0) 
print(mean_m) ### ici la dimension colonne (2) disparait (on ne garde que les lignes)
print()
mean_m = np.mean(m,axis = 1)
print(mean_m)
print()
mean_c = np.mean(m,axis = None)
print(mean_c)
```

### ça marche avec plein d'opérateur

```python
### std
std_c_1 = np.std(m,axis = 1)

print(std_c_1)
```

```python
### max
max_c_0 = np.max(m,axis = 0)

print(max_c_0)
```

```python
### min
min_c_0 = np.min(m,axis = 0)

print(min_c_0)
```

```python
### argmax
argmax_c_0 = np.argmax(m,axis = 0)

print(argmax_c_0)
```

### alegbre lineaire

```python
a = np.ones(shape = (4,4))*3
b = np.ones(shape = (4,4))*2
print(a)
print(b)
```

```python
### produit de chaque element

c = a*b 
print(c)
```

```python
# produit matricielle

c = np.dot(a,b)
print(c)
```

```python
c = a @ b
print(c)
```

```python
## idem pour inverse

mat = rng.integers(low=1, high=6, size=(4,4)).astype("float64")

print(mat)
```

```python
# division par element
c = 1 / mat
print(c)
```

```python
## on obtient la matrice inverse
inv_mat = np.linalg.inv(mat)

print(inv_mat)
```

```python
np.round((inv_mat @ mat))
```
