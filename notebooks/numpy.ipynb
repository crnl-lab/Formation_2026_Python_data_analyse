# Exercice 2
Voici un vecteur
a = np.random.randn(5000)


Créer un vecteur b qui enlève les déviants.

- \> M + 2.5*std
- < M - 2.5*std

note : np.mean(a) donne la moyenne de a et np.std(a) donne l'écart-type de a




````python
import numpy as np
rng = np.random.default_rng(seed=None)
a = rng.normal(size=5000, loc=13, scale=5)

m = np.mean(a)
s = np.std(a)

print(a.shape)
print(m,s)

keep = (a>(m-2.5*s)) & (a<(m+2.5*s))
print(keep)
a2 = a[keep]
print(a2.shape)
b = a[~keep]
print(b.shape)
````

### Exercice 3: pseudo potentiel evoqué moyen

A partie de fausse data : 
  * Signal de 20 canaux (= électrodes) de 10000 points temporels (Gaussien) chacun
  * De **triggers** : `trigs = np.array([100,500,2000,2500,3500,4000,6000,7500,8600,9000])`

Il faut decouper:

  1. le signal en tronçon autrour de chaque trigger de -5 points à +10 sample
  2. Empiler tout ça dans un array 3d
  3. Faire la moyenne et std sur les trials



````python
trigs = np.array([100,500,2000,2500,3500,4000,6000,7500,8600,9000])
trigs.size
print(trigs.shape)

rng = np.random.default_rng(seed=2205)
sigs = rng.normal(size=(10_000, 20))
print(sigs.shape)
````

````python
stacked_sigs = np.zeros((15, 20, trigs.size))
print(stacked_sigs.shape)


for i in range(trigs.size):
    trig = trigs[i]
#     print(i, trig)
#     print(sigs[trig-5:trig+10, :].shape)
    stacked_sigs[:, :, i] = sigs[trig-5:trig+10, :]

m = np.mean(stacked_sigs, axis=2)
s = np.std(stacked_sigs, axis=2)
print(m.shape)
````

````python
# Solution 2 : elegant mais pas efficasse en mémoire

stacked_sigs =[]
for trig in trigs:
    stacked_sigs.append(sigs[trig-5:trig+10, :])
    
stacked_sigs = np.stack(stacked_sigs, axis = 2)
print(stacked_sigs.shape)
````

![image](img/np_exo_argmin.png)


````python
courbes = np.zeros((200, 5))
t = np.arange(200) / 20.
for i in range(5):
    courbes[:, i] = t**3 - (8 + i) * t**2 - t +5
````

````python
latences = np.argmin(courbes, axis=0)
print(latences)
min_values = courbes[m, np.arange(5)]
print(min_values)
````

````python
# teaser sur matplotlib
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(courbes)
m = np.argmin(courbes, axis=0)
ax.scatter(latences, min_values, color='black')
````
